Goal: Write a **function** that will plot $f(x) = x^2$ and $g(x) = \sqrt{x}$ on the same plot and return the $f(x)$ and $g(x)$ values as separate lists. Your plot needs to have the following:
- An appropriate figure size (`figsize=(10,8)` is recommended)
- Labeled x and y axis
- A title
- Different colors and linestyles for each function
- A legend
- Your domain should be $x \in [0, 10]$. *Hint: Use `np.linspace` to set a range of values for your functions.*

Your function should return:
- A list of $f(x)$ values
- A list of $g(x)$ values.

```python
# Graph!
def plot_functions():  # Define function to plot values -- inputs are x = 0 to 5, outputs are x^2 or sqrt(x)
    x = np.linspace(0, 10)  # Domain
    
    f_x = x ** 2  # f(x)
    g_x = x ** (1/2)  # g(x)
    
    plt.figure(figsize=(10, 8))  # Figure size (according to recommendation)

    plt.title('Plot of f(x) = x^2 and g(x) = sqrt(x)', fontsize=18)  # Title of plot
    plt.xlabel('x-axis', fontsize=14)  # Labeling x-axis
    plt.ylabel('y-axis', fontsize=14)  # Labeling y-axis

    plt.plot(x, f_x, label='f(x) = x^2', color='blue', linestyle='--')  # Creating label, color, and linestyle of f(x)
    plt.plot(x, g_x, label='g(x) = sqrt(x)', color='green', linestyle='-')  # Creating label, color, and linestyle of g(x)

    plt.legend()  # Legend
    plt.show()  # Show graph

    return f_x.tolist(), g_x.tolist()  # This line changes the numpy arrays of f(x) and g(x) as lists, thus when it is called upon, a list will be returned

f_values, g_values = plot_functions()  # Show plot with both functions and return f(x) and g(x) as lists
```

```python
# My Function!
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from matplotlib.ticker import LinearLocator
from matplotlib.lines import Line2D  # To make legend

fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10, 8))  # Create a figure with a larger size and a 3D subplot

X = np.linspace(-3, 3, 100)  # Uses a finer grid to get a smoother surface
Y = np.linspace(-3, 3, 100)  # Uses a finer grid to get a smoother surface
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X ** 2 + Y ** 2)
Z = np.sin(R) + np.cos(2 * R)  # Surface function

surf = ax.plot_surface(X, Y, Z, cmap=cm.viridis,  # Changed colormap
                       edgecolor='none',  # No gridlines
                       alpha=0.9)  # Add slight transparency, to see the whole graph

ax.set_zlim(-2, 2)  # Add a color bar for the surface plot
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter('{x:.02f}')

cbar = fig.colorbar(surf, shrink=0.7, aspect=12)  # Creates a colorbar for the z-axis, showing 'elevation'
cbar.set_label('Z-value Intensity', fontsize=12)

ax.set_xlabel('X-axis', fontsize=12, labelpad=10)  # Add axis labels and a title
ax.set_ylabel('Y-axis', fontsize=12, labelpad=10)
ax.set_zlabel('Z-axis', fontsize=12, labelpad=10)
ax.set_title('Coolest 3D Surface Plot Ever', fontsize=16)

ax.view_init(elev=30, azim=135)  # Rotate the view

legend_proxy = Line2D([0], [0], color='green', label='Z = sin(R) + cos(2R)')  # Legend
ax.legend(handles=[legend_proxy], loc='upper left', fontsize=12)

plt.show()  # Show plot
```
